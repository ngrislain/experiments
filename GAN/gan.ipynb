# Quick GAN experiment

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
%matplotlib inline
```

```python
# Data generatioln,
N = 1000
theta = 7*np.random.randn(N)
X = np.cos(theta) + 0.2*np.random.randn(N)
Y = np.sin(theta) + 0.2*np.random.randn(N)

plt.plot(X,Y,'ro',alpha=0.2)
```
