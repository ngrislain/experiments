```python
import math
import numpy as np
# A class
class E(object):
    # Utility functions
    @staticmethod
    def unpair(i,j):
        return (i+j)*(i+j+1)/2+j
    @staticmethod
    def pair(k):
        if k!=None and k>=0:
            w = int((math.sqrt(8*k+1)-1)/2)
            t = (w**2+w)/2
            return (w+t-k, k-t)
        else:
            return (None, None)
    # The constructor
    def __init__(self, i, j=None):
        if i!=None:
            if j!=None:
                self.i = i
                self.j = j
                self.k = E.unpair(i,j)+1
            else:
                self.k = i
                self.i, self.j = E.pair(self.k-1)
        else:
            self.i = None
            self.j = None
            self.k = 0
    # A multiplication
    def __mul__(self, e):
        if self.i!=None and e.j!=None and self.j==e.i:
            return E(self.i, e.j)
        else:
            return E(None, None)
    def dense(self, n):
        result = np.zeros((n,n))
        if self.k!=None:
            result[self.i, self.j] = 1
        return result
    def __repr__(self):
        if self.k!=None:
            return 'e_{},{}'.format(self.i, self.j)
        else:
            return '0'
```

```python
e = E(1,2)*E(2,3)
print e.dense(5)
```

```python
e = E(2,3)
print e.k
print e.i, e.j
print E(19)
print E(33)
print E(19)*E(33)
print E(100000)
print (E(18)*E(32)).k
```

```python
#matplotlib inline
import matplotlib.pyplot as plt

N = 1000
A = np.zeros((N,N))
for i in xrange(N):
    for j in xrange(N):
        A[i,j] = (E(i)*E(j)).k

plt.imshow(A)
plt.show()
```

# A new implementation

```python
import math
import numpy as np
# A class
class E(object):
    N = 17
    # Utility functions
    @staticmethod
    def unpair(i,j):
        return i+E.N*j
    @staticmethod
    def pair(k):
        if k!=None and k>=0:
            return (k%E.N, k/E.N)
        else:
            return (None, None)
    # The constructor
    def __init__(self, i, j=None):
        if i!=None:
            if j!=None:
                self.i = i
                self.j = j
                self.k = E.unpair(i,j)+1
            else:
                self.k = i
                self.i, self.j = E.pair(self.k-1)
        else:
            self.i = None
            self.j = None
            self.k = 0
    # A multiplication
    def __mul__(self, e):
        if self.i!=None and e.j!=None and self.j==e.i:
            return E(self.i, e.j)
        else:
            return E(None, None)
    def dense(self):
        result = np.zeros((E.N,E.N))
        if self.k!=None:
            result[self.i, self.j] = 1
        return result
    def __repr__(self):
        if self.k!=None:
            return 'e_{},{}'.format(self.i, self.j)
        else:
            return '0'
```

```python
e = E(1,2)*E(2,3)
print e.dense()
```

```python
e = E(2,3)
print e.k
print e.i, e.j
print E(19)
print E(24)
print E(19)*E(24)
print E(10)
print (E(19)*E(24)).k
```

```python
#matplotlib inline
import matplotlib.pyplot as plt

N = E.N*E.N+1
A = np.zeros((N,N))
for i in xrange(N):
    for j in xrange(N):
        A[i,j] = (E(i)*E(j)).k

plt.imshow(A)
plt.show()
```

# A tensor approach

```python
import numpy as np
import numpy.linalg as la

N = 4
T = np.zeros((N,N,N))
for i in xrange(N):
    for j in xrange(N):
            T[i,j,i] = 1

def mat(T, n):
    N = len(T.shape)
    d = T.shape[n]
    D = np.prod(T.shape)/d
    M = T.transpose([(i+n)%N for i in xrange(N)]).reshape(d, D)
    return np.dot(M,M.T)

print mat(T, 0)
print
print mat(T, 1)
print
print mat(T, 2)

np.tensordot(T, T, 0)
```
